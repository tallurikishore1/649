```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/Instagram_visits.csv')
df
df.head()
```

```python
print("Missing values:\n", df.isnull().sum())
df.dropna(inplace=True)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df[['Instagram visit score', 'Spending_rank(0 to 100)']])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Instagram visit score'], df['Spending_rank(0 to 100)'], c=df['Cluster'], cmap='viridis', alpha=0.6)
plt.xlabel('Instagram Visit Score')
plt.ylabel('Spending Rank (0 to 100)')
plt.title('User Clusters based on Visit Score and Spending Rank')
plt.colorbar(label='Cluster')
plt.show()
```
